# Data preprocessing
1 lower case
2 tokenization
3 removing special character
4 removing stop words and punctuation
stemming


```python
# the whole function to do the preprocessing
```

```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    
    y=[]
    for i in text:
        if i.isalnum():
            y.append(i)
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
             y.append(i)
    
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
from nltk.corpus import stopwords
stopwords.words('english') # stopwords
```

```python
import string
string.punctuation
```

```python
df['text'][2001]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
ps.stem('loving')
```

```python
df['transform_text'] = df['text'].apply(transform_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X = cv.fit_transform(df['transform_text']).toarray()   #toarray = to convert the sparse array to dense one
```

```python
X.shape
```

```python
y = df['target'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix, precision_score
```

```python
#object creation
gaussNB = GaussianNB()
multiNB = MultinomialNB()
bernNB = BernoulliNB()
```

```python
gaussNB.fit(X_train,y_train)
y_pred1 = gaussNB.predict(X_test)
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
```

```python
multiNB.fit(X_train,y_train)
y_pred2 = multiNB.predict(X_test)
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
```

```python
bernNB.fit(X_train,y_train)
y_pred3 = bernNB.predict(X_test)
print(accuracy_score(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
print(precision_score(y_test,y_pred3))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=3000)
```

```python
X = tfidf.fit_transform(df['transform_text']).toarray()
```

```python
y = df['target'].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix, precision_score
```

```python
gaussNB = GaussianNB()
multiNB = MultinomialNB()
bernNB = BernoulliNB()
```

```python
gaussNB.fit(X_train,y_train)
y_pred1 = gaussNB.predict(X_test)
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
```

```python
multiNB.fit(X_train,y_train)
y_pred2 = multiNB.predict(X_test)
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
```

```python
bernNB.fit(X_train, y_train)
y_pred3 = bernNB.predict(X_test)
print(accuracy_score(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
print(precision_score(y_test,y_pred3))
```
